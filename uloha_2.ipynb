```python
#Dependencies
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from plotly.subplots import make_subplots
df = pd.read_csv('/home/ladi/Documents/msw_zapocet/GHSH_Pooled_Data1.csv')
```

```python
#Graph 1
df_mean = df.groupby('Country')[['Attempted_suicide']].mean().sort_values('Attempted_suicide',ascending = False)
fig_bar = px.bar(x = df_mean.index, y = df_mean.Attempted_suicide, color = df_mean.Attempted_suicide, color_continuous_scale='dense')
fig_bar.update_layout(width = 850, height = 500, title_text="suicide rates by Country")
fig_bar.update_xaxes(title = 'Country')
fig_bar.update_yaxes(title = 'Suicide rate')
fig_bar.show()
```

```python
#Graph 2
overweight = df.sort_values(by='Overwieght', ascending=True)
sns.set_theme(style="whitegrid")
plt.figure(figsize=(40,15))
sns.boxplot(x = 'Country', y = 'Overwieght', data = overweight)
plt.xlabel('Country', fontsize=40)
plt.ylabel('Overweight', fontsize=40)
plt.xticks(rotation = 45)
plt.title('Suicide rates in corellation with overweight issues.', fontsize=60, color='black')
```

```python
#Graph 3
use_majoranka = df.sort_values(by='Use_Marijuana', ascending=False)
plt.figure(figsize=(60,15))
sns.lineplot(x = 'Country', y = 'Use_Marijuana', data = use_majoranka)
plt.xlabel('Country', fontsize=20)
plt.ylabel('Percentage of teenagers using marijuana.', fontsize=20)
plt.title('Teenagers using marijuana according to age.', fontsize=30, color='black')
```

```python
#Graph 4
year = df.sort_values(by='Year', ascending=True)
plt.figure(figsize=(60,15))
sns.pointplot(x = 'Year', y = 'Attempted_suicide', data = year)
plt.xlabel('Year', fontsize=20)
plt.ylabel('Attempted suicide rate', fontsize=20)
plt.title('Attempted suicides in corresponding years.', fontsize=30, color='black')
```

```python
#Graph 5
fig = px.pie(df, values='Attempted_suicide', names='Year', title='Percentage of people attempting suicide by year.')
fig.show()
```
