```python
#Dependencies
import numpy as np
import matplotlib.pyplot as plt
import timeit
```

```python
def bisection(f, a0, b0, eps=1E-10):
    a, b = a0, b0
    while b - a > 2*eps:
        x = (a + b)/2
        if f(a)*f(x) < 0:
            b = x
        else:
            a = x
    return x

def newton(f, a0, b0, delta=1E-10):
    a, b = a0, b0
    def df(x, h = 1E-10):
        return (f(x+h)-f(x-h))/(2*h)
    xnew = (a+b)/2
    xold = a
    while abs(xnew - xold) > delta:
        xold = xnew
        xnew = xold - f(xold)/df(xold)
    return xnew

def plot_result(f, a0, b0, x, title):
    print("root: ", x)
    plt.title(title)
    plt.plot(x, f(x), "bo")                                              
    plt.plot(np.linspace(a0, b0, 100), f(np.linspace(a0, b0, 100)), "r")
    plt.axhline(y=0)
    plt.show()
```

```python
fce1 = lambda x: x**2 + 5*x - 8
fce2 = lambda x: 3*x - np.exp(4*x) + 15
fce3 = lambda x: np.sin(x-2)
```

```python
#Function 1

a, b = -2 , 4
plot_result(fce1, a, b, bisection(fce1, a, b), "Bisection: -x^3 + 3x^2 + x - 3")
plot_result(fce1, a, b, newton(fce1, a, b), "Newton: -x^3 + 3x^2 + x - 3")
timeb = timeit.timeit(lambda: bisection(fce1, a, b), number=15000)
timen1 = timeit.timeit(lambda: newton(fce1, a, b), number=15000)

print(f"BISECTION: {timeb/15000}s, NEWTON: {timen1/15000}s")
```

```python
#Function 2
a, b = -2, 2
plot_result(fce2, a, b, bisection(fce2, a, b), "Bisection: 8x - e^(-5x) + 5")
plot_result(fce2, a, b, newton(fce2, a, b), "Newton: 8x - e^(-5x) + 5")
timeb = timeit.timeit(lambda: bisection(fce2, a, b), number=15000)
timen2 = timeit.timeit(lambda: newton(fce2, a, b), number=15000)

print(f"BISECTION: {timeb/15000}s, NEWTON: {timen2/15000}s")
```
