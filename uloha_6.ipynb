```python
#Dependencies
from turtle import position
import pyautogui
import time
import numpy as np
import random as r
positions = []
seed = 0
```

```python
print('Move your mouse randomly around the screen for next 5 seconds to create random seed.')
for i in range(25):
    x, y = pyautogui.position()
    positions.append((x, y))
    print(f'Mouse position '+str(x) + ' '+str(y) + ' saved.')
    for i in positions:
        seed += np.math.factorial(sum(i))
    time.sleep(0.2)
print(f'Your randomly generated seed is: {seed}')
```
